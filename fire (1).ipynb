## Multicollinaerity for categorical_feature

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in cat_feature:
    if chi2_contingency(pd.crosstab(df['Classes  '], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[cat_feature, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

### we are going to apply the linear regression 


```python
df.head()
```

```python
x = df.drop(columns =["Temperature"])
y = df.Temperature
```

#### spliting the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=10)
```

```python
## Standardize or feature scaling the datasets
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

#### model building using linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

#### finding the coefficent and the intercept 

```python
print("coefficent of the model is :",lr.coef_)
print("intercept of the model is:",lr.intercept_)
```

#### PRediction for the test data

```python
lr_pred=lr.predict(x_test)
```

```python
lr_pred
```

#### validation of the assumption

```python
## 1) variation of predicted and the expected result has linear variation
sns.scatterplot(y_test,lr_pred)
plt.xlabel("expected result")
plt.ylabel("predicted result")
```

```python
##2)  residual variation should be normally distributed
residual = y_test-lr_pred
sns.kdeplot(residual)
### there are more outliers in the left direction and it is not good model
```

```python
## SCatter plot with predictions and residual should be uniform distributed
plt.scatter(lr_pred,residual)
### so the condition of the homoscedasticity is satisfied
```

```python
## Performance Metrics
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print("mean squre error of the model is :",mean_squared_error(y_test,lr_pred))
print("mean absoulte error of the model is:",mean_absolute_error(y_test,lr_pred))
print("mean root error of the model is:",np.sqrt(mean_squared_error(y_test,lr_pred)))
```

### R square and adjusted R square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,lr_pred)
print(score)
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
